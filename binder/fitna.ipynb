Prepare data

```python
import numpy as np
import fitna

np.random.seed(12345)

# Two dimensional normal distributions
normal_params = [
    fitna.data.NormalDist(norm=1, mean=np.array([4.5, 4.5]), cov=np.array([[1, -0.5], [-0.5, 1]]), size=100),
    fitna.data.NormalDist(norm=1, mean=np.array([6.0, 7.0]), cov=np.array([[1,  0.5], [ 0.5, 1]]), size=100)
]

normal_mixture = fitna.data.NormalMixture(normal_params)
data_points = normal_mixture.rvs()
```

Plot the data

```python
import plotly

plotly.offline.init_notebook_mode(connected=True)

# Make traces with data points
data = []

for point_set in data_points:
    trace = plotly.graph_objs.Scatter(
        x = point_set[0],
        y = point_set[1],
        mode = 'markers',
        hoverinfo = 'none'
    )
    data.append(trace)


# Make traces with covariance matrix eigenvectors
initial_estimates = [
    fitna.data.NormalDist(norm=0.5, mean=np.array([4.0, 8.0]), cov=np.array([[1., 0.], [0., 1.]]) ),
    fitna.data.NormalDist(norm=0.5, mean=np.array([4.01, 8.01]), cov=np.array([[1., 0.], [0., 1.]]) )
]

data.extend(fitna.plotly.make_traces(initial_estimates))


# Create a slider for step-by-step animation
slider = dict( active = 0, steps  = [] )

slider_step = { 'args': [ { 'data': data } ], 'method': 'animate' }
slider['steps'].append(slider_step)


# Now add other steps
data_points_T = np.concatenate(data_points, 1).T
ll_new, all_estimates = fitna.em.do_em(data_points_T, initial_estimates, 1e-5, 50)
    
for estimates in all_estimates:
    err_traces = fitna.plotly.make_traces(estimates)
    #del data[-2:]
    data = data[:-2]
    data.extend(err_traces)
    
    slider_step = { 'args': [ { 'data': data } ], 'method': 'animate' }
    slider['steps'].append(slider_step)


layout = {'height': 1000, 'yaxis': {'scaleanchor': "x", 'scaleratio': 1}}

layout['sliders'] = [slider]

figure = plotly.graph_objs.Figure(data=data, layout=layout)

plotly.offline.iplot(figure, show_link=False, link_text="")
```
